### Create a function to generate prompts based on two parameters: text and generateExplanation.
#### If generateExplanation is set to True, include an explanation; if set to False, no explanation will be provided.

```python
def getSentiment(text, generateExplanation):
    #Check if explanation generation is enabled
    if generateExplanation:
        #If enabled, create a prompt instructing the model to classify sentiment and provide an explanation
        prompt= f"Please classify the sentiment in the following text as either 'fear', 'anger' or 'joy' and provide some explanation.\n\n{text}'\n\nSentiment: "
    else:
        #If explanation generation is not enabled, create a prompt for sentiment classification only
        prompt= f"Please classify the sentiment in the following text as either 'fear', 'anger' or 'joy'.\n\n{text}'\n\nSentiment: "
    
    #Use the OpenAI GPT-3.5 Turbo model to generate a response based on the provided prompt
    response = client.chat.completions.create(
      model="gpt-3.5-turbo", # Specify the GPT-3.5 Turbo model
      messages=[
        {"role": "system", "content": prompt} # Include the system-level prompt
     
  ]
)
 
    #Extract and return the content of the first choice in the response
    return response.choices[0].message.content
```

```python
#load and read data
sent_pred_path = Path("./Resource/Emotion_classify_Data.csv")
# Read file 
sent_prediction_df = pd.read_csv(sent_pred_path)

sentiment_prediction_df = sent_prediction_df.copy()
```

```python
#display dataframe
sent_prediction_df
```

## STEP1: Makeing predictions without propmt engineering.

```python
#split the test and train set from the datarafram
N = len(sent_prediction_df)
N_train = int( .8 * N )
N_test = N - N_train

test_set = sent_prediction_df[-N_test:]
```

```python
#Initialize empty lists to store predicted sentiments and ground truth (GT) sentiments
total = 0
correct = 0

predicted = [] #Predicted sentiments
gt = [] #Ground truth sentiments

#Iterate through the rows in the test_set
for i,row in test_set.iterrows():
    
    #Get the sentiment prediction using the getSentiment function (explanation generation is disabled)
    sentiment = getSentiment(row['Comment'],False)
    
    #Append the ground truth sentiment from the DataFrame to the 'gt' list
    gt.append( row['Emotion'] )
    #Append the predicted sentiment to the 'predicted' list
    predicted.append(sentiment)
            
    #Check if the predicted sentiment matches the ground truth sentiment 
    #Increment the 'total' count to keep track of the total number of predictions
    total+=1
    if sentiment == row['Emotion']:
        correct+=1
       
    #Print the comment, ground truth sentiment, and predicted sentiment for analysis
    print( row['Comment'], '===>  GT: ' , row['Emotion'],  ' Predicted', sentiment)
    print('---------------------------------------------------------------------')
    
    

```

### Calculate the confusion matrix and plot it without propmt engineering and based just 3 emotions

```python
#dependecy for confusion matrix
from sklearn.metrics import confusion_matrix
```

```python
# Define a list of labels that represent sentiment categories
labels = ['fear', 'joy', 'anger']
# Compute the confusion matrix using ground truth (gt) and predicted sentiment values
cf = confusion_matrix(gt, predicted, labels = labels)
```

```python
#display confusion matrix
cf
```

```python
#dependency to plot ConfusionMatrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
#plot Confusion Metrix before count other sentiments
ConfusionMatrixDisplay(cf, display_labels=labels).plot() 
```

## STEP2:  In this section, we request the prompt engineering to make predictions specifically based on the three initial sentiments: fear, anger, and joy.

```python
text = "I woke up on a beautiful sunday morning feeling restless and miserable"
prompt= f"Please classify the sentiment in the following text as either 'fear', 'anger' or 'joy'. make sure the predicted sentiment is only one of the previously stated emotions.\n\n{text}'\n\nSentiment: "

#Import the OpenAI library and any other necessary dependencies
from openai import OpenAI

#Create an OpenAI client by providing an API key
client = OpenAI(
    # Defaults to os.environ.get("OPENAI_API_KEY")
     api_key=openAI_key
)

#Define a system-level prompt or instruction that will guide the AI model
response = client.chat.completions.create(
  model="gpt-3.5-turbo", # Specify the GPT-3.5 Turbo model
  messages=[
    {"role": "system", "content": prompt}  # System-level instruction or prompt
     
  ]
)

print(response.choices[0].message.content)
```

```python
def getSentiment_v2(text, generateExplanation):
    #Check if explanation generation is enabled
    if generateExplanation:
        #If enabled, create a prompt instructing the model to classify sentiment and provide an explanation
        prompt= f"Please classify the sentiment in the following text as either 'fear', 'anger' or 'joy' and provide some explanation.\n\n{text}'\n\nSentiment: "
    else:
        
        #If explanation generation is not enabled, create a prompt for sentiment classification only
        prompt= f"Please classify the sentiment in the following text as either 'fear', 'anger' or 'joy'.Make sure the predicted sentiment is only one of the previously stated emotions.\n\n{text}'\n\nSentiment: "
    
    #Use the OpenAI GPT-3.5 Turbo model to generate a response based on the provided prompt
    response = client.chat.completions.create(
      model="gpt-3.5-turbo", # Specify the GPT-3.5 Turbo model
      messages=[
        {"role": "system", "content": prompt} # Include the system-level prompt
     
  ]
)
 
    #Extract and return the content of the first choice in the response
    return response.choices[0].message.content
```

```python
#Initialize empty lists to store predicted sentiments and ground truth (GT) sentiments

total_2 = 0
correct_2 = 0

predicted_2 = [] #Predicted sentiments
gt_2 = [] #Ground truth sentiments

#Iterate through the rows in the test_set
for i,row in test_set.iterrows():
    
    #Get the sentiment prediction using the getSentiment function (explanation generation is disabled)
    sentiment = getSentiment_v2(row['Comment'],False)
    
    #Append the ground truth sentiment from the DataFrame to the 'gt' list
    gt_2.append( row['Emotion'] )
    #Append the predicted sentiment to the 'predicted' list
    predicted_2.append(sentiment)
            
    #Check if the predicted sentiment matches the ground truth sentiment 
    #Increment the 'total' count to keep track of the total number of predictions
    total_2+=1
    if sentiment == row['Emotion']:
        correct_2+=1
       
    #Print the comment, ground truth sentiment, and predicted sentiment for analysis
    print( row['Comment'], '===>  GT: ' , row['Emotion'],  ' Predicted', sentiment)
    print('---------------------------------------------------------------------')
    
    

```

```python
cf_2= confusion_matrix(gt_2, predicted_2, labels = labels_extended)
#display the cf_2
cf_2
```
